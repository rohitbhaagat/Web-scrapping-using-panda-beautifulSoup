```python
# Step 5: Find all rows in the table body
rows = table_body.find_all('tr')
rows
```

```python
# Step 6: Extract data from each row
column_info = []
for row in rows:
    cols = row.find_all('td')
    country_info = [col.text.strip() for col in cols]  # Strip whitespace characters
    column_info.append(country_info)
```

```python
# Step 7: Create lists to hold each column's data
Number = []
Country = []
GDP_norminal = []
GDP_abbrev = []
GDP_growth = []
Population = []
GDP_per_capital = []
Share_of_world_GDP = []
```

```python
# Step 8: Populate the lists with data
for i in column_info:
    Number.append(i[0])
    Country.append(i[1])
    GDP_norminal.append(i[2])
    GDP_abbrev.append(i[3])
    GDP_growth.append(i[4])
    Population.append(i[5])
    GDP_per_capital.append(i[6])
    Share_of_world_GDP.append(i[7])
```

```python
# Step 9: Create a dictionary with the lists
Data = {
    'Number': Number,
    'Country': Country,
    'GDP nominal': GDP_norminal,
    'GDP abbrev': GDP_abbrev,
    'GDP growth': GDP_growth,
    'Population': Population,
    'GDP per capita': GDP_per_capital,
    'Share of world GDP': Share_of_world_GDP
}
```

```python
# Step 10: Convert the dictionary to a DataFrame
df = pd.DataFrame(Data)
```

```python
# Step 11: Save the DataFrame to a CSV file
df.to_csv('gdp_by_country.csv', index=False)

print("Data has been saved to gdp_by_country.csv")
```
